```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris

from sklearn.cluster import KMeans as SK_KMeans
```

```python
from K_Means_Scratch import KMeans_Scratch
```

## **Data**

```python
data = load_iris()
```

```python
X = data.data
```

## **Parametros**

```python
n_clusters = 3
max_iter = 300
random_state = 13
```

## **Comparación de los modelos**

### **KMeans Scratch**

```python
# Entremanmos nuestro modelo
kmeans_scratch = KMeans_Scratch(n_clusters=3, max_iter=300, random_state=13)
kmeans_scratch.fit(X)
```

```python
kmeans_scratch.labels
```

### **KMeans Scikit-Learn**

```python
# Entrenamos el modelo de Scikit-learn 
sklearn_kmeans = SK_KMeans(n_clusters=3, random_state=42)
sklearn_kmeans.fit(X)
```

### **Comparación**

```python
fig, ax = plt.subplots(1,2, figsize=(14, 4))

ax[0].scatter(X[:, 0], X[:, 1], c=kmeans_scratch.labels)
ax[0].scatter(
    kmeans_scratch.clusters_center[:, 0], 
    kmeans_scratch.clusters_center[:, 1], 
    s=100, 
    c='red', 
    label='Centroids', 
    marker='x'
)
ax[0].set_title('Kmeans Scratch')

ax[1].scatter(X[:, 0], X[:, 1], c=sklearn_kmeans.labels_)
ax[1].scatter(
    sklearn_kmeans.cluster_centers_[:, 0], 
    sklearn_kmeans.cluster_centers_[:, 1], 
    s=100, 
    c='red', 
    label='Centroids', 
    marker='x'
)
ax[1].set_title('Kmeans Scikit-learn')

plt.show()
```
