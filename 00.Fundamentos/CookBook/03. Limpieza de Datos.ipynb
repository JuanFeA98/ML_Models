## 3.18 Concatenar Dataframes

```python
data_a = {
    'id': ['1', '2', '3'],
    'first': ['Sandra', 'Billy', 'Allen'],
    'last': ['Martínez', 'Ackerman', 'Ali']
}

data_b = {
    'id': ['4', '5', '6'],
    'first': ['Alex', 'Amy', 'Bran'],
    'last': ['Bonder', 'Balwer', 'Black']
}
```

```python
df_a = pd.DataFrame(data_a, columns=['id', 'first', 'last'])
df_b = pd.DataFrame(data_b, columns=['id', 'first', 'last'])
```

```python
# Concatenar por filas

pd.concat([df_a, df_b], axis=0)
```

```python
# Concatenar por columnas

pd.concat([df_a, df_b], axis=1)
```

```python
# Agregando una fila

new_row = pd.Series([10, 'Chris', 'Chillon'], index=['id', 'first', 'last'])
```

```python
df_a.append(new_row, ignore_index=True)
```

## 3.19 Uniendo Dataframes

```python
datos = {
    'id': ['1', '2', '3', '4'],
    'name': ['Amy Jones', 'Allen Keys', 'Alice Bees', 'Tim Horton']
}

ventas = {
    'id': ['3', '4', '5', '6'],
    'ventas': [23456, 2512, 2345, 1455]
}
```

```python
df_emp = pd.DataFrame(datos)
df_ventas = pd.DataFrame(ventas)
```

```python
# Merge tipo inner

pd.merge(df_emp, df_ventas, on='id')
```

```python
# Merge tipo outer

pd.merge(df_emp, df_ventas, on='id', how='outer')
```
